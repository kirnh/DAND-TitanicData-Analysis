# Analysis of Titanic data.                                                                        

#### Purpose:
Analysing the data to understand the factors that affected the survival rate of people on the titanic.

#### Process:
First the data was imported into a pandas dataframe. It was then explored in many ways where the survival rate was considered as the dependent variable with independent variables being the different groups of people on the titanic. 

The groups were based on either gender, class of travel or age group. Deeper exploration was conducted by further dividing people into groups based on gender and class of travel, age group and class of travel or gender and age group. Grouping of people based on all three of the independent variables being gender, class of travel and age group was also made. And the survival rate for each of those groups was calculated for further analysis. 

#### Code:
The below code section of the report contains the stages of data exploration and data wrangling that was conducted to get insights into the titanic data with our purpose in mind.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
#importing data from csv into a pandas dataframe
titanic_df = pd.read_csv('titanic_data.csv')
```

```python
#looking into the format of our data
titanic_df.head()
```

```python

#function that returns percentage of people survived in a dataframe 
def percent_survived(df):
    return df['Survived'].mean() * 100.0

#function that returns the count of people survived and not survived in a group
def count_people(df):
    survived = df['Survived'].sum()
    not_survived = df['Survived'].count() - survived
    return survived, not_survived
```

```python
#grouping data by Sex
passengers_by_gender = titanic_df.groupby('Sex')

#plotting the survival rate according to gender
print passengers_by_gender.apply(percent_survived)
passengers_by_gender.apply(percent_survived).plot(kind = 'bar')
plt.show()
```

```python
#grouping data by class
passengers_by_class = titanic_df.groupby('Pclass')

#plotting survival rate by class
print passengers_by_class.apply(percent_survived)
passengers_by_class.apply(percent_survived).plot(kind = 'bar')
plt.show()
```

```python
#creating a new column to show age group (child, teen, adult or senior)
#to create age group data, 'Age' column entry with NaN values are dropped
#Age_group category definitions are as follows:
#    child: age 0 - age 12
#    teen: age 13 - age 19
#    adult: age 20 - age 59
#    senior: age 60 above (considering 120 as the maximum)

titanic_df['Age_group'] = pd.cut(titanic_df['Age'].dropna(), bins = [0,13,19,60,120], \
                                 labels = ['child', 'teen', 'adult', 'senior'])

#grouping data by age group
passengers_by_age_group = titanic_df.groupby('Age_group')

#plotting survival rate by age group
print passengers_by_age_group.apply(percent_survived)
passengers_by_age_group.apply(percent_survived).plot(kind = 'bar')
plt.show()
```

```python
#grouping data by gender and Pclass
passengers_by_gender_and_class = titanic_df.groupby(['Sex','Pclass'])

#plotting survival rate by gender and Pclass
print passengers_by_gender_and_class.apply(percent_survived)
passengers_by_gender_and_class.apply(percent_survived).plot(kind = 'bar')
plt.show()
```

```python
#grouping data by Age_group and Pclass
passengers_by_age_group_and_class = titanic_df.groupby(['Age_group', 'Pclass'])

#plotting survival rate by Age_group and Pclass
print passengers_by_age_group_and_class.apply(percent_survived)
passengers_by_age_group_and_class.apply(percent_survived).plot(kind = 'bar')
plt.show()
```

```python
#grouping data by Sex and Age_group
passengers_by_gender_and_age_group = titanic_df.groupby(['Sex', 'Age_group'])

#plotting survival rate by Sex and Age_group
print passengers_by_gender_and_age_group.apply(percent_survived)
passengers_by_gender_and_age_group.apply(percent_survived).plot(kind = 'bar')
plt.show()
```

```python
#grouping data by Sex, Pclass and Age_group
passengers_by_gender_class_and_age_group = titanic_df.groupby(['Sex', 'Pclass', 'Age_group'])

#plotting survival rate by Sex, Pclass and Age_group
print passengers_by_gender_class_and_age_group.apply(percent_survived)
passengers_by_gender_class_and_age_group.apply(percent_survived).plot(kind = 'bar')
plt.show()
```

#### Questionzing the purpose:

Question that satisfies our purpose of understanding the factors that affected the survival rate of people on titanic can be formed as, 'What are the characteristics of people that made them more likely to survive the titanic sinking?'

The above explorations show that females, children and first class travellers in general survived the titanic tragedy better.

However on top of that, some groups have a survival rate of 100%. These groups are as follows:
    - Children in second class.
    - Female seniors.
    - Female teenagers in first class and second class.
    - Male children in first class.

A survival rate of 100% is interesting and requires some more investigation.
