```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df = pd.read_csv('titanic_data.csv')
titanic_df.head()
```

```python
titanic_df.describe()
```

```python
#grouping data by sex
passengers_by_gender = titanic_df.groupby('Sex')
passengers_by_gender.describe()
```

```python
#grouping data by class
passengers_by_class = titanic_df.groupby('Pclass')
passengers_by_class.describe()
```

```python
#adding a column of age group to the original dataframe (data with NaN values for Age column have been dropped)
titanic_df['Age_group'] = pd.cut(titanic_df['Age'].dropna(), bins = [0, 13, 19, 60, 120], labels = ['child', 'teen', 'adult', 'senior'])
```

```python
#grouping data by age_group
passengers_by_age_group = titanic_df.groupby('Age_group')
passengers_by_age_group.describe()
```

```python
#function that returns the percentage of females in a 
def percent_females(df):
    f = 0 
    for i in df['Sex']:
        if i == 'female':
            f += 1
    return f * 100.0 / len(df['Sex'])

#function that returns percentage of people survived in a dataframe 
def percent_survived(df):
    return df['Survived'].sum() * 100.0 / df['Survived'].count()

#function that returns the count of people survived and not survived in a group
def count_people(df):
    survived = df['Survived'].sum()
    not_survived = df['Survived'].count() - survived
    return survived, not_survived
```

```python
print passengers_by_class.apply(percent_survived)
print passengers_by_age_group.apply(percent_survived)
print passengers_by_gender.apply(percent_survived)
```

```python
#grouping data by Pclass and Sex
passengers_by_class_and_gender = titanic_df.groupby(['Pclass', 'Sex'])

#calculating the percentage of people survived
passengers_by_class_and_gender.apply(percent_survived)
```

```python
#grouping data by Pclass and Age_group
passengers_by_class_and_age_group = titanic_df.groupby(['Age_group', 'Pclass'])

#calculating the percentage of people survived
passengers_by_class_and_age_group.apply(percent_survived)
```

```python
#grouping data by Age_group and Sex
passengers_by_age_group_and_gender = titanic_df.groupby(['Age_group', 'Sex'])

#calculating the percentage of people survived
passengers_by_age_group_and_gender.apply(percent_survived)
```
